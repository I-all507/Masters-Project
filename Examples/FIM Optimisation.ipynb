# Optimising h #
This notebook contains a function for optimising the step sizes for the 6-parameter Fisher matrix, parametrised by $\{ \log \mathcal{A}, \log\mathcal{M}_z, t_c,\phi_c,\eta,z\}$. The optimisation works by treating the six step sizes as a vector, $\vec h$, and finding the position in step space that minimises the distance between analytical and numerical results of the IFIM.

```python
z = 0.01
detector = 'ET'
params = [8,2,proper_distance(z), 10.424*np.pi/180, 43.631*np.pi/180,0,0,'APR4'] # Starting with an optimally oriented system wrt the ET.

_, err_analyt = get_analyt(params,z,detector) # get_analyt() retrives the analytic IFIM uncertainties for a given NSBH system.

h0 = np.array([1.77223575e-10,1.82213789e-10,1.77223575e-10,1.52947750e-9, 1e-10,1e-2]) # Guessing a random starting point in step-space.

print('Running optimisation...')
errors_vector, h_opt = optimiser(analytical_errors=err_analyt, params=params, h0=h0,yapper=True,h_limit=1e-12,z=0.01,detector='ET',score_limit=5e-8)
```

Quickly check the difference in errors...

```python
print(f'Numerical errors: {errors_vector}, \n Analytical errors: {err_analyt}')
print(f'Fraction deviations: {errors_vector/err_analyt}')
```

That's pretty accurate! But to account for any randomness, and any potential local minima in step space, the optimiser should be ran several times.

```python
h_vals = []
errors = []

print('Running optimisation...')
for i in range(1,101):
    print(f'test: {i}')
    h0 = np.array([10**-11, 10**-11, 10**-11, 10**-11,10**-11,10**-6])*np.random.rand(6)*1000 # Generating random entry points for optimisation process.
    
    err, h_opt = optimiser(analytical_errors=err_analyt, params=params, h0=h0,yapper=False,h_limit=1e-12,z=z,detector=detector,score_limit=5e-8) ## Setting score to be reasonably low
                                                                                                                                                 ## so optimisation doesn't last forever.
    errors.append(err)
    h_vals.append(h_opt)
```

# (1) Analysis #
With 100 optimisations ran, the distribution of $\vec{h}$ components and error estimates can be observed. Let's find the configuration that minimised the score and plot that over the results of optimisation.

```python
def get_score(vector,log_node):
    def difference(point, log_node=log_node):
        return np.abs(log_node-point)
    return np.dot(difference(point=vector),difference(point=vector))
x = []

for i in errors:
    x.append(get_score(np.log(i),log_node=np.log(err_analyt)))
x=np.array(x)
print('The optimal h values for the PN waveform are: ',h_vals[x.argmin()])
print('This had a score of: ',x[x.argmin()])
print('The errors associated with this h vector are: ', errors[x.argmin()])
print('And the fractional deviation is: ', errors[x.argmin()]/err_analyt)
h_best = h_vals[x.argmin()]
best_dev = errors[x.argmin()]/err_analyt
```
